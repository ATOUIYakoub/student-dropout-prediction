```python
# notebooks/1_eda.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# Load data
df = pd.read_csv('../data/student-mat.csv', sep=';')

# Basic info
print(df.shape)
print(df.info())
print(df.describe())

```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Dropout proxy: Low final grade or many absences
df['dropout'] = ((df['G3'] < 10) & (df['absences'] > 10)).astype(int)

```

```python
# Visualize dropout vs some features
sns.countplot(x='dropout', data=df)
plt.title("Dropout Distribution")
plt.show()
```

```python

# Grades vs dropout
sns.boxplot(x='dropout', y='G3', data=df)
plt.title("Final Grade vs Dropout")
plt.show()

```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
